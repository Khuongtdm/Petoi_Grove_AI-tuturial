
#### Validate YOLO Dataset Structure

For YOLO object detection, it's essential that each image has a corresponding label file (`.txt`) with the same base filename in the *same directory*. This function helps to validate this structure, especially when dealing with nested folders.

```
import os
if flag ==3:
  def validate_yolo_dataset(root_dir, image_extensions=('.jpg', '.jpeg', '.png', '.bmp', '.tiff')):
      """
      Validates a YOLO dataset laid out as two parallel folder trees:

          root_dir/
              images/  (optionally with subfolders like train/ val/ test/)
              labels/  (same subfolder structure, .txt files)

      Matches files by their path *relative to* the images/ or labels/ folder,
      with the extension stripped -- so images/train/dog.jpg pairs with
      labels/train/dog.txt even though they live in different directories.

      Returns:
          dict with:
            'valid_pairs': list of (image_path, label_path)
            'images_without_labels': list of image paths missing a label
            'labels_without_images': list of label paths missing an image
            'num_images_found': int
            'num_labels_found': int
            'images_dir': resolved path or None
            'labels_dir': resolved path or None
      """
      # Find the images/ and labels/ dirs (case-insensitive, anywhere under root_dir,
      # in case there's an extra nesting level like root_dir/dataset/images)
      images_dir = None
      labels_dir = None
      for dirpath, dirnames, _ in os.walk(root_dir):
          for d in dirnames:
              low = d.lower()
              if low == 'images' and images_dir is None:
                  images_dir = os.path.join(dirpath, d)
              elif low == 'labels' and labels_dir is None:
                  labels_dir = os.path.join(dirpath, d)
          if images_dir and labels_dir:
              break

      result = {
          'valid_pairs': [],
          'images_without_labels': [],
          'labels_without_images': [],
          'num_images_found': 0,
          'num_labels_found': 0,
          'images_dir': images_dir,
          'labels_dir': labels_dir,
      }

      if not images_dir or not labels_dir:
          result['error'] = (
              f"Could not find both 'images' and 'labels' folders under {root_dir}. "
              f"Found images_dir={images_dir}, labels_dir={labels_dir}"
          )
          return result

      # Build a map: relative path (no extension) -> full image path
      image_map = {}
      for dirpath, _, filenames in os.walk(images_dir):
          for f in filenames:
              base, ext = os.path.splitext(f)
              if ext.lower() in image_extensions:
                  rel_dir = os.path.relpath(dirpath, images_dir)
                  key = os.path.normpath(os.path.join(rel_dir, base))
                  image_map[key] = os.path.join(dirpath, f)

      # Build a map: relative path (no extension) -> full label path
      label_map = {}
      for dirpath, _, filenames in os.walk(labels_dir):
          for f in filenames:
              base, ext = os.path.splitext(f)
              if ext.lower() == '.txt':
                  rel_dir = os.path.relpath(dirpath, labels_dir)
                  key = os.path.normpath(os.path.join(rel_dir, base))
                  label_map[key] = os.path.join(dirpath, f)

      result['num_images_found'] = len(image_map)
      result['num_labels_found'] = len(label_map)

      all_keys = set(image_map) | set(label_map)
      for key in all_keys:
          img = image_map.get(key)
          lbl = label_map.get(key)
          if img and lbl:
              result['valid_pairs'].append((img, lbl))
          elif img and not lbl:
              result['images_without_labels'].append(img)
          elif lbl and not img:
              result['labels_without_images'].append(lbl)

      return result


  if __name__ == "__main__":
      # Point this at your dataset root (the folder that contains images/ and labels/)
      kaggle_base_dir = path  # 'path' should already hold the dataset download location
      content_items = os.listdir(kaggle_base_dir)
      actual_kaggle_dataset_root = kaggle_base_dir
      if len(content_items) == 1 and os.path.isdir(os.path.join(kaggle_base_dir, content_items[0])):
          actual_kaggle_dataset_root = os.path.join(kaggle_base_dir, content_items[0])

      print(f"Analyzing dataset structure at: {actual_kaggle_dataset_root}")
      results = validate_yolo_dataset(actual_kaggle_dataset_root)

      if 'error' in results:
          print(f"\nERROR: {results['error']}")
      else:
          print(f"\nimages/ dir: {results['images_dir']}")
          print(f"labels/ dir: {results['labels_dir']}")
          print(f"Total images found: {results['num_images_found']}")
          print(f"Total label files found: {results['num_labels_found']}")
          print(f"Valid image-label pairs found: {len(results['valid_pairs'])}")

          if results['images_without_labels']:
              print(f"\nImages without labels ({len(results['images_without_labels'])} total, showing 5):")
              for p in results['images_without_labels'][:5]:
                  print(f"  - {p}")

          if results['labels_without_images']:
              print(f"\nLabels without images ({len(results['labels_without_images'])} total, showing 5):")
              for p in results['labels_without_images'][:5]:
                  print(f"  - {p}")

          if results['valid_pairs']:
              print("\nConclusion: Dataset appears to be a valid YOLO images/labels dataset.")
          else:
              print("\nConclusion: No valid image-label pairs found.")
```

#### Organize Kaggle Dataset and Generate `data_kaggle.yaml` for YOLOv8


```
import os
import shutil
import random
import yaml

# ---------------------------------------------------------------
# 1. Use dataset already downloaded via kagglehub in a previous cell
#    (expects `path` variable to already exist, e.g.:
#       path = kagglehub.dataset_download("mustafaxgm/classroom-student-engagement-dataset")
# ---------------------------------------------------------------
assert 'path' in dir(), "Run the kagglehub.dataset_download(...) cell first to define `path`."
print(f"Using Kaggle dataset at: {path}")

kaggle_output_base_dir = './split_data_kaggle'
VAL_SPLIT = 0.2  # 20% of data goes to validation
SEED = 42

kaggle_base_dir = path
content_items = os.listdir(kaggle_base_dir)
actual_kaggle_dataset_root = kaggle_base_dir
if len(content_items) == 1 and os.path.isdir(os.path.join(kaggle_base_dir, content_items[0])):
    actual_kaggle_dataset_root = os.path.join(kaggle_base_dir, content_items[0])

kaggle_source_root = actual_kaggle_dataset_root
print(f"Processing Kaggle dataset from resolved root: {kaggle_source_root}")

source_images_dir = os.path.join(kaggle_source_root, 'images')
source_labels_dir = os.path.join(kaggle_source_root, 'labels')

# ---------------------------------------------------------------
# 2. Create target structure: images/{train,val}, labels/{train,val}
# ---------------------------------------------------------------
for modality in ['images', 'labels']:
    for split in ['train', 'val']:
        os.makedirs(os.path.join(kaggle_output_base_dir, modality, split), exist_ok=True)

# ---------------------------------------------------------------
# 3. Match image/label pairs, then split
# ---------------------------------------------------------------
IMAGE_EXTS = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')

if not os.path.exists(source_images_dir):
    print(f"Error: Source images directory not found at {source_images_dir}. Cannot proceed.")
elif not os.path.exists(source_labels_dir):
    print(f"Error: Source labels directory not found at {source_labels_dir}. Cannot proceed.")
else:
    # Build stem -> filename maps
    image_files = {os.path.splitext(f)[0]: f for f in os.listdir(source_images_dir)
                    if os.path.splitext(f)[1].lower() in IMAGE_EXTS}
    label_files = {os.path.splitext(f)[0]: f for f in os.listdir(source_labels_dir)
                   if f.endswith('.txt')}

    paired_stems = sorted(set(image_files) & set(label_files))
    unmatched_images = set(image_files) - set(label_files)
    unmatched_labels = set(label_files) - set(image_files)

    print(f"Found {len(paired_stems)} valid image-label pairs.")
    if unmatched_images:
        print(f"  Warning: {len(unmatched_images)} images have no matching label (skipped).")
    if unmatched_labels:
        print(f"  Warning: {len(unmatched_labels)} labels have no matching image (skipped).")

    random.seed(SEED)
    random.shuffle(paired_stems)
    val_count = max(1, int(len(paired_stems) * VAL_SPLIT))
    val_stems = set(paired_stems[:val_count])
    train_stems = set(paired_stems[val_count:])

    print(f"Splitting: {len(train_stems)} train / {len(val_stems)} val")

    def copy_pair(stem, split):
        img_name = image_files[stem]
        lbl_name = label_files[stem]
        shutil.copy2(
            os.path.join(source_images_dir, img_name),
            os.path.join(kaggle_output_base_dir, 'images', split, img_name)
        )
        shutil.copy2(
            os.path.join(source_labels_dir, lbl_name),
            os.path.join(kaggle_output_base_dir, 'labels', split, lbl_name)
        )

    for stem in train_stems:
        copy_pair(stem, 'train')
    for stem in val_stems:
        copy_pair(stem, 'val')

    print(f"Kaggle dataset organized and copied to {kaggle_output_base_dir}")

# ---------------------------------------------------------------
# 4. Infer class names
# ---------------------------------------------------------------
class_names_kaggle = []
num_classes_kaggle = 0
kaggle_data_yaml_source = os.path.join(kaggle_source_root, 'data.yaml')

if os.path.exists(kaggle_data_yaml_source):
    print(f"Found data.yaml in Kaggle dataset at: {kaggle_data_yaml_source}")
    with open(kaggle_data_yaml_source, 'r') as f:
        kaggle_config = yaml.safe_load(f)
        if 'names' in kaggle_config:
            class_names_kaggle = kaggle_config['names']
            num_classes_kaggle = len(class_names_kaggle)
            print(f"Extracted {num_classes_kaggle} class names from data.yaml: {class_names_kaggle}")
        else:
            print(f"Warning: 'names' key not found in {kaggle_data_yaml_source}. Will try to infer from labels.")

if not class_names_kaggle:
    print("Attempting to infer class names from label files in the organized dataset...")
    all_class_ids = set()
    train_labels_dir = os.path.join(kaggle_output_base_dir, 'labels', 'train')
    if os.path.exists(train_labels_dir):
        for root, _, files in os.walk(train_labels_dir):
            for file in files:
                if file.endswith('.txt'):
                    with open(os.path.join(root, file), 'r') as f:
                        for line in f:
                            try:
                                class_id = int(line.strip().split(' ')[0])
                                all_class_ids.add(class_id)
                            except (ValueError, IndexError):
                                continue
    if all_class_ids:
        num_classes_kaggle = max(all_class_ids) + 1
        class_names_kaggle = [f'class_{i}' for i in range(num_classes_kaggle)]
        print(f"Inferred {num_classes_kaggle} classes from label files. Using placeholder names: {class_names_kaggle}")
    else:
        print("Could not infer class names. Setting default 1 class 'object'.")
        num_classes_kaggle = 1
        class_names_kaggle = ['object']

# ---------------------------------------------------------------
# 5. Write data_kaggle.yaml (paths match images/{train,val} layout)
# ---------------------------------------------------------------
data_kaggle_yaml_content = f"""train: {kaggle_output_base_dir}/images/train
val: {kaggle_output_base_dir}/images/val

nc: {num_classes_kaggle}  # number of classes
names: {class_names_kaggle} # class names
"""

kaggle_yaml_file_path = './data_kaggle.yaml'
with open(kaggle_yaml_file_path, 'w') as f:
    f.write(data_kaggle_yaml_content)

print(f"Generated YOLOv8 data_kaggle.yaml at: {kaggle_yaml_file_path}")
print("Content of data_kaggle.yaml:")
print(data_kaggle_yaml_content)

# ---------------------------------------------------------------
# 6. Clean up original download
#    NOTE: this deletes the kagglehub cache dir referenced by `path`.
#    If you want to re-run this cell later (e.g. with a different
#    VAL_SPLIT), you'll need to re-run the kagglehub download cell
#    first, or comment this block out.
# ---------------------------------------------------------------
print(f"Removing original Kaggle dataset from {kaggle_base_dir}...")
try:
    shutil.rmtree(kaggle_base_dir)
    print("Original Kaggle dataset removed.")
except FileNotFoundError:
    print(f"Original Kaggle dataset directory {kaggle_base_dir} already not found. Skipping removal.")
except Exception as e:
    print(f"Error removing original Kaggle dataset: {e}")

flag = 3
print(f"Set flag = {flag} for Kaggle dataset training.")
```
